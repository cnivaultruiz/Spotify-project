```python
import pandas as pd
from time import sleep

from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import pickle
# transformer?
# other preprocessing?
```

```python
tracks_and_features_df = pd.read_csv('tracks_and_features.csv')
features_df = pd.read_csv('features.csv')
```

```python
audio_features_model_on = ['is_explicit','popularity', 'danceability', 'energy', 'key','loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'tempo', 'time_signature']
features_df = features_df[audio_features_model_on]
```

```python
audio_features_model_on_A=['is_explicit','popularity', 'danceability', 'energy', 'key',
       'loudness', 'mode', 'speechiness']
audio_features_model_on_B = [ 'acousticness', 'instrumentalness',
       'liveness', 'tempo', 'time_signature']
features_df_A = tracks_and_features_df[audio_features_model_on_A]
features_df_B = tracks_and_features_df[audio_features_model_on_B]
```

```python
features_df
```

```python
km100 = KMeans(100)
km200_A= KMeans(100)
km200_B= KMeans(100)
```

```python
km100
```

```python
scaler = MinMaxScaler()
scaler_A = MinMaxScaler()
scaler_B = MinMaxScaler()


scaled_features = scaler.fit_transform(features_df)
km100.fit(scaled_features)

scaled_features_A = scaler_A.fit_transform(features_df_A)
km200_A.fit(scaled_features_A)

scaled_features_B = scaler_B.fit_transform(features_df_B)
km200_B.fit(scaled_features_B)
```

```python
tracks_clustered_df = tracks_and_features_df.copy()
tracks_clustered_df['cluster_km100'] = km100.labels_
tracks_clustered_df['cluster_km200_A'] = km200_A.labels_
tracks_clustered_df['cluster_km200_B'] = km200_B.labels_

tracks_clustered_df[tracks_clustered_df['cluster_km100']==1].sample(1)
```

```python
tracks_clustered_df.to_csv('tracks_clustered_df.csv', index=False) # save our dataset with cluster info
# save the model with pickle
# save any transformers pickle


# loading and saving with pickle:

# import pickle

# a = {'hello': 'world'}

# saving/dumping
# with open('filename.pickle', 'wb') as handle:
#     pickle.dump(a, handle, protocol=pickle.HIGHEST_PROTOCOL)

# loading
# with open('filename.pickle', 'rb') as handle:
#     b = pickle.load(handle)

# print(a == b)

with open('model_km100.pickle', 'wb') as handle:
    pickle.dump(km100, handle, protocol=pickle.HIGHEST_PROTOCOL)
    pickle.dump(km200_A, handle, protocol=pickle.HIGHEST_PROTOCOL)
    pickle.dump(km200_B, handle, protocol=pickle.HIGHEST_PROTOCOL)


with open('scaler.pickle', 'wb') as handle:
    pickle.dump(scaler, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open('scaler_A.pickle', 'wb') as handle:
    pickle.dump(scaler_A, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open('scaler_B.pickle', 'wb') as handle:
    pickle.dump(scaler_B, handle, protocol=pickle.HIGHEST_PROTOCOL)

display(tracks_clustered_df)
```
